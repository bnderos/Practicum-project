<font size="4">**Описание данных:**</font>     

**Технологический процесс**

+ Rougher feed — исходное сырье
+ Rougher additions (или reagent additions) — флотационные реагенты: Xanthate, Sulphate, Depressant
    + Xanthate — ксантогенат (промотер, или активатор флотации);
    + Sulphate — сульфат (на данном производстве сульфид натрия);
    + Depressant — депрессант (силикат натрия).
+ Rougher process (англ. «грубый процесс») — флотация
+ Rougher tails — отвальные хвосты
+ Float banks — флотационная установка
+ Cleaner process — очистка
+ Rougher Au — черновой концентрат золота
+ Final Au — финальный концентрат золота

**Параметры этапов**
+ air amount — объём воздуха
+ fluid levels — уровень жидкости
+ feed size — размер гранул сырья
+ feed rate — скорость подачи

Наименование признаков должно быть такое:
[этап].[тип_параметра].[название_параметра]

+ rougher — флотация

+ secondary_cleaner — вторичная очистка
+ final — финальные характеристики

**Возможные значения для блока [тип_параметра]:**

+ input — параметры сырья
+ output — параметры продукта
+ state — параметры, характеризующие текущее состояние этапа
+ calculation — расчётные характеристики

### Эффективность обогащения рассчитана правильно?

**Вычислим её на обучающей выборке для признака rougher.output.recovery. Найдём MAE между нашими расчётами и значением признака.** 

```python
# напишем функцию для расчета эффективности обогащения
def recover(c, f, t):
    rec = (c * (f - t)) / (f * (c - t)) * 100
    return rec
```

```python
recovery = recover(df_train['rougher.output.concentrate_au'],
                  df_train['rougher.input.feed_au'], 
                   df_train['rougher.output.tail_au']
                  )
```

**У нас есть выбросы в таргетах, удалим их.**

```python
# напишем функцию для удаления выбросов через расчёт квантилей
def quant(df, col):
    q1 = df[col].quantile(.25)  # первый квартиль Q1
    q3 = df[col].quantile(.75)  # третий квартиль Q3
    iqr = q3 - q1                 # межквартильный размах IQR
    df = df[(df[col] > (q1 - 1.5 * iqr)) & (df[col] < (q3 + 1.5 * iqr))]
    return df
```

```python
df_train = quant(df_train, Y[0])
df_train = quant(df_train, Y[1])
df_test = quant(df_test, Y[0])
df_test = quant(df_test, Y[1])
```

### Концентрация металлов (Au, Ag, Pb) на различных этапах

**Посмотрим, как меняется концентрация металлов (Au, Ag, Pb) на различных этапах: в сырье, в черновом концентрате, в концентрате после первой очистки и в финальном концентрате.**

```python
# напишем функцию для графиков
def graphics_conc(lst_cycle, lst_label, met):
    plt.figure(figsize = (20, 8))
    for i in lst_cycle:
        sns.distplot(df_full[i], hist=False, kde_kws = {'shade': True, 'linewidth': 3})
        plt.legend(lst_label)
    plt.title(f'Концентрация {met} на различных этапах', size=25)
    plt.xlabel(f'Концентрация {met}', size=20)
    plt.grid(True)
```

```python
%%time
lst_label = ['сырьё',
             'черновой концентрат',
             'концентрат после первой очистки',
             'финальный концентрат']
# создадим таблицу для будущих значений
metals = pd.DataFrame(columns=lst_label)
for metal in ['au', 'ag', 'pb']:
    # создадим список этап + металл
    lst_step = [STEP[item]+metal for item in range(len(STEP))]
    # заполним таблицу средними значениями каждого этапа для каждого металла
    metals.loc[metal] = [df_full[m].mean() for m in lst_step]
    # построим график для каждого металла
    graphics_conc(lst_step, lst_label, metal)
```

```python
# посмотрим на средние значения каждого этапа для каждого металла
metals
```

**Вывод:**  
+ концентрация золота, как мы видим, с каждым этапом становится значительно выше.  
+ содержание серебра сначала растёт, затем начитает падать   
+ концентрация свинца постепенно повышается, на стадиях "после первой очистки" и "финальный концентрат" почти одинаковое. 
+ также мы видим, что присутствуют значительное количество нулевых значений.

## Модель

### Напишем функцию для вычисления итоговой sMAPE

```python
def smape(y, pred):
    s = abs(y - pred) / ((abs(y) + abs(pred))/2) * 100
    return s.sum() / len(y)
```

```python
def res_smape(y, pred):
    y_rougher = y[Y[0]]
    y_final = y[Y[1]]
    pred_rougher = pred[:, 0]
    pred_final = pred[:, 1]
    return .25 * smape(y_rougher, pred_rougher) + .75 * smape(y_final, pred_final)
```

### Обучим разные модели и оценим их качество кросс-валидацией.

```python
df_train = df_train.reset_index(drop=True)
df_test = df_test.reset_index(drop=True)
```

```python
# разделим фичи и таргет на трейне и тесте
X_train = df_train.drop(Y, axis=1)
y_train = df_train[Y]
X_test = df_test.drop(Y, axis=1)
y_test = df_test[Y]
```

```python
X_train.shape, y_train.shape, X_test.shape, y_test.shape
```

```python
# создадим scorer для проверки качества
scorer_smape = make_scorer(res_smape, greater_is_better=False)
```

**Масштабирование признаков мы делать не будем. Подбор гиперпараметров и кроссвалидацию будем делать при помощи GridSearchCV.**

#### Дерево решений

```python
%%time
grid_par_DT = {'max_depth': range(2, 11)}
model_DT = DecisionTreeRegressor(random_state=RANDOM_STATE)
cv_DT = GridSearchCV(model_DT, grid_par_DT, scoring=scorer_smape)
cv_DT.fit(X_train, y_train)
sMAPE = -cv_DT.best_score_
print(f'Лучший параметр: {cv_DT.best_params_}\nЛучший показатель sMAPE: {sMAPE:.6f}')
```

#### Линейная регрессия

```python
%%time
model_LR = LinearRegression()
model_LR.fit(X_train, y_train)
sMAPE = -cross_val_score(model_LR, X_train, y_train, cv=5, scoring=scorer_smape).mean()
print(f'Лучшее значение sMAPE для линейной регрессии: {sMAPE:.6f}')
```

#### Случайный лес

```python
%%time
grid_par_RF = {'max_depth': [3, 5], 
              'n_estimators': [20, 30]}
model_RF = RandomForestRegressor(random_state=RANDOM_STATE)
cv_RF = GridSearchCV(model_RF, grid_par_RF, scoring=scorer_smape)
cv_RF.fit(X_train, y_train)
sMAPE = -cv_RF.best_score_
print(f'Лучший параметр: {cv_RF.best_params_}\nЛучший показатель sMAPE: {sMAPE:.6f}')
```

#### CatBoost

```python
def res_smape_CB(rougher, final):
    return .25 * rougher + .75 * final
```

```python
scorer_smape_CB = make_scorer(smape, greater_is_better=False)
```

```python
%%time
grid_par_CB = {'depth': [6, 8],
            'learning_rate': [0.05, 0.1],
            'iterations': [30, 100]}
model_CB_rougher = GridSearchCV(CatBoostRegressor(random_state=RANDOM_STATE), grid_par_CB, scoring=scorer_smape_CB)
model_CB_rougher.fit(X_train, y_train[Y[0]], verbose=False)
sMAPE_rougher = -model_CB_rougher.best_score_
model_CB_final = GridSearchCV(CatBoostRegressor(random_state=RANDOM_STATE), grid_par_CB, scoring=scorer_smape_CB)
model_CB_final.fit(X_train, y_train[Y[1]], verbose=False)
sMAPE_final = -model_CB_final.best_score_
sMAPE = res_smape_CB(sMAPE_rougher, sMAPE_final)
print(f'Лучший параметр rougher: {model_CB_rougher.best_params_}\n'
      f'Лучший параметр final:{model_CB_final.best_params_}\n'
      f'Лучший показатель sMAPE: {sMAPE:.6f}')
```

**Лучший sMAPE показала модель CatBoost, проверим её на тестовой выборке.**

```python
model_CB_rougher = CatBoostRegressor(random_state=RANDOM_STATE,
                                     depth=6,
                                     learning_rate=0.05,
                                     iterations=100)
model_CB_rougher.fit(X_train, y_train[Y[0]], verbose=False)
pred_CB_rougher = model_CB_rougher.predict(X_test)
sMAPE_rougher = smape(y_test[Y[0]], pred_CB_rougher)
print(f'Лучший показатель sMAPE: {sMAPE_rougher:.6f}')
model_CB_final = CatBoostRegressor(random_state=RANDOM_STATE,
                                     depth=6,
                                     learning_rate=0.1,
                                     iterations=30)
model_CB_final.fit(X_train, y_train[Y[1]], verbose=False)
pred_CB_final = model_CB_final.predict(X_test)
sMAPE_final = smape(y_test[Y[1]], pred_CB_final)
print(f'Лучший показатель sMAPE: {sMAPE_final:.6f}')
sMAPE = res_smape_CB(sMAPE_rougher, sMAPE_final)
print(f'Лучший показатель sMAPE: {sMAPE:.6f}')
```
