### Разобьём данные на выборки

**Данные разобьём на три части: обучающую, валидационную и тестовую. Размеры тестового и валидационного наборов обычно равны. Исходные данные разобьём в соотношении 3:1:1.**

```python
rand = 123
# разобьём данные на признаки и целевой признак
X = df_ohe.drop(['exited'], axis=1)
y = df_ohe['exited']
```

```python
X.shape, y.shape
```

```python
# разобьём сначала на обучающий и тестовый наборы
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=rand)
```

```python
# затем разобьём на тестовый и валидационный наборы
X_test, X_val, y_test, y_val = train_test_split(X_test, y_test, test_size=0.5, random_state=rand)
```

```python
# проверим длины наших выборок
X_train.shape, y_train.shape, X_test.shape, y_test.shape, X_val.shape, y_val.shape
```

**Мы разбили датасет на признаки, а такжа на обучающую, валидационную и тестовую выборки.**

### Масштабирование признаков

**Стандартизируем численные признаки. Создадим объект структуры StandardScaler() и настроем его на обучающих данных.**

```python
df_ohe.nunique()
```

```python
numeric = ['credit_score', 'age', 'tenure', 'balance', 'num_of_products', 'estimated_salary']
```

```python
scaler = StandardScaler()
scaler.fit(X_train[numeric])
X_train[numeric] = scaler.transform(X_train[numeric])
X_val[numeric] = scaler.transform(X_val[numeric])
X_test[numeric] = scaler.transform(X_test[numeric])
X_train.head()
```

```python
X_val.head()
```

```python
X_test.head()
```

### Обучение моделей без учёта дисбаланса

**Посмотрим на частоты классов целевого признака.**

```python
class_frequency = y.value_counts(normalize=True)
print(class_frequency)
class_frequency.plot(kind='bar');
```

#### Дерево решений

```python
# подберём оптимальные гиперпараметры
best_model_DT = None
best_depth_DT = 0
best_res_DT = 0
lst_depth = []
lst_f1_val = []
lst_f1_tr = []
for depth in tqdm(range(1, 11)):
    model_DT = DecisionTreeClassifier(random_state=rand, max_depth=depth)  # создаём модель
    model_DT.fit(X_train, y_train)                                        # обучаем её
    predict_DT_val = model_DT.predict(X_val)
    f1_DT = f1_score(y_val, predict_DT_val)
    f1_DT_tr = f1_score(y_train, model_DT.predict(X_train))
    lst_depth.append(depth)
    lst_f1_val.append(f1_DT)
    lst_f1_tr.append(f1_DT_tr)
    if f1_DT > best_res_DT:
        best_res_DT = f1_DT
        best_model_DT = model_DT
        best_depth_DT = depth
print(f'F1-мера наилучшей модели: {best_res_DT:.4f}\nМаксимальная глубина: {best_depth_DT}')
```

У модели "дерево решений" наилучшие показатели **при глубине 6 - 0.5849**. Обучим на ней модель и посчитаем метрики.

```python
model_DT = DecisionTreeClassifier(random_state=rand, max_depth=6)  # создаём модель
model_DT.fit(X_train, y_train)                                        # обучаем её
predict_DT_val = model_DT.predict(X_val)
```

```python
# создадим функцию для метрик
def quality_metrics(model, X_val, y, pred):
    print(f'Показатели accuracy = {accuracy_score(y, pred):.4f}')
    print(f'Полнота = {recall_score(y, pred):.4f}')
    print(f'Точность = {precision_score(y, pred):.4f}')
    print(f'F-1 мера = {f1_score(y, pred):.4f}')
    prob_val = model.predict_proba(X_val)
    print('AUC_ROC =', round(roc_auc_score(y, prob_val[:, 1]), 4))
```

```python
quality_metrics(model_DT, X_val, y_val, predict_DT_val)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(lst_depth, lst_f1_val, 'o-')
plt.plot(lst_depth, lst_f1_tr, 'o-')
plt.grid(True)
plt.legend(['F1 val', 'F1 train'])
plt.title('F1-мера в зависимости от глубины', fontsize=20)
plt.xlabel('Глубина', fontsize=15)
plt.ylabel('F1-мера', fontsize=15);
```

**Построим матрицу ошибок для дерева решений.**

```python
cm = confusion_matrix(y_val, predict_DT_val)
ConfusionMatrixDisplay(confusion_matrix=cm).plot();
```

**По показателям метрик и матрице можно сказать, что модель склонна выдавать больше отрицательных предсказаний. Довольно низкий показатель полноты, модель плохо ищет долю TP-ответов среди всех, у которых истинная метка 1. Это предсказуемо, учитывая дисбаланс классов.**

#### Случайный лес

```python
def RF(X, y, cl_w):
# подберём оптимальные гиперпараметры
    best_model_RF = None
    best_f1_RF = 0
    best_est = 0
    best_depth_RF = 0
    for est in tqdm(range(30, 81, 10)):
        for depth in range(5, 16):
            model_RF = RandomForestClassifier(random_state=rand, n_estimators=est, max_depth=depth, class_weight=cl_w)
            model_RF.fit(X, y)
            predict_RF_val = model_RF.predict(X_val)
            f1_RF = f1_score(y_val, predict_RF_val)
            if f1_RF > best_f1_RF:
                best_model_RF = model_RF
                best_f1_RF = f1_RF
                best_est = est
                best_depth_RF = depth
    print(f'F1 наилучшей модели: {best_f1_RF:.4f} \nКоличество деревьев: {best_est} \nМаксимальная глубина: {best_depth_RF}')
```

```python
RF(X_train, y_train, None)
```

У модели "случайный лес" наилучшие показатели при количестве деревьев **40** и глубине **9** - **0.6006**. Случайный лес работает довольно медленно: чем больше деревьев, тем неторопливее работает модель. Обучим на них модель, посчитаем метрики и построим матрицу ошибок.

Создадим функцию upsample() с тремя параметрами:
+ X — признаки,
+ y — целевой признак,
+ repeat — количество повторений.    

Функция вернёт признаки и целевой признак после операции upsampling.

```python
def upsample(X, y, repeat):
    X_zeros = X[y == 0]
    X_ones = X[y == 1]
    y_zeros = y[y == 0]
    y_ones = y[y == 1]
    
    X_upsampled = pd.concat([X_zeros] + [X_ones] * repeat)
    y_upsampled = pd.concat([y_zeros] + [y_ones] * repeat)
    
    X_upsampled, y_upsampled = shuffle(X_upsampled, y_upsampled, random_state=rand)
    return X_upsampled, y_upsampled
```

###  Уменьшение выборки

Напишем функцию downsample() с тремя аргументами:
+ features — признаки;
+ target — целевой признак;
+ fraction — доля отрицательных объектов, которые нужно сохранить.

Функция вернёт признаки и целевой признак после операции downsampling.

```python
def downsampling(X, y, fraction):
    X_zeros = X[y == 0]
    X_ones = X[y == 1]
    y_zeros = y[y == 0]
    y_ones = y[y == 1]
    
    X_down = pd.concat([X_zeros.sample(frac=fraction, random_state=rand)] + [X_ones])
    y_down = pd.concat([y_zeros.sample(frac=fraction, random_state=rand)] + [y_ones])
    
    X_down, y_down = shuffle(X_down, y_down, random_state=123)
    return X_down, y_down
```
